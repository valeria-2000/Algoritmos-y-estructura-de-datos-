```
import requests 
url = 'https://listado.mercadolibre.com.co/calzado-tenis/'
response = requests.get (url)
print (response)
```

```
import bs4
parseo = bs4.BeautifulSoup (response.text, 'html.parser')
getDataSelect = lambda clase : parseo.select(clase)
dataTitle = parseo.select ('.ui-search-item__title')
dataPrice = getDataSelect('.price-tag-fraction')
print (dataTitle)
print (dataPrice)
elementosTennis = []
funcionExtraerTexto = lambda elemento : elemento.text
funcionListaDatos = lambda funcion, data :  list (map (funcion,data))
elementosTennis = funcionListaDatos(funcionExtraerTexto, dataTitle)
print (elementosTennis)

listaPrecios = funcionListaDatos(funcionExtraerTexto, dataPrice)
print (listaPrecios)
```

```
print(len(elementosTennis))
print(len(listaPrecios))
listaPrecios = listaPrecios[:51]
print(len(listaPrecios))
```

```
import pandas as pd
dictTenis = {}
dictTenis['Titulos'] = elementosTennis
print (len (elementosTennis), len (listaPrecios))
dictTenis ['Precios'] = listaPrecios
tenisDataFrame = pd.DataFrame(dictTenis)
tenisDataFrame.to_csv('Tenis.csv', index=False)
```

```
funcionAgregarTenis = lambda elemento : 'Tenis : ' + elemento
titulosTenis = list (map(funcionAgregarTenis,elementosTennis))
print (titulosTenis)
titulosTenis = funcionListaDatos (funcionAgregarTenis, elementosTennis)
print (titulosTenis )
dictTenis = {}
dictTenis['Titulos'] = titulosTenis 
dictTenis ['Precios'] = listaPrecios
TenisDF = pd.DataFrame(dictTenis)
TenisDF.to_csv('TenisMod.csv', index= False) 
```

```
caballero = lambda elemento : 'Caballero' in elemento 

dama = lambda elemento : 'Dama' in elemento 

listacaballero = list (filter(caballero, titulosTenis ))
print (listacaballero)

listadama = list (filter(dama, titulosTenis))
print (listadama)

fitSerie = pd.Series(listacaballero)
fitSerie.to_csv('Caballero.csv', index_label = 'Caballero')

damaSerie = pd.Series(listadama)
damaSerie.to_csv ('Dama.csv', index_label = 'Dama')
```

```
from functools import reduce 
titulosTenis =  list (elementosTennis)
sumador = lambda acumulado = 0, elemento = 0: acumulado  + elemento 
suma_elementos = reduce (sumador, titulosTenis)
print (suma_elementos)
```

```
titulosTenis = list (elementosTennis)
nombresTenisCortos = lambda elemento : len (elemento) < 25
listaNombresFiltrados = list ( filter(nombresTenisCortos, titulosTenis))
print (listaNombresFiltrados)
```

```
from functools import reduce 

lista = titulosTenis
size = lambda elemento: len (elemento)

listaSize = list (map(size,lista))
sumador = lambda elemento = 0, acumulador = 0 : acumulador + elemento 
promedio = reduce (sumador, listaSize) / len( titulosTenis)
print (promedio)
```
